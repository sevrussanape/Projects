```python
import os
import random
import shutil

data_folder = 'data'
summary_folder = 'summary'
output_folder = 'output'

# Set the seed for random shuffling
random.seed(42)

# Create output folders
train_folder = os.path.join(output_folder, 'train')
test_folder = os.path.join(output_folder, 'test')
valid_folder = os.path.join(output_folder, 'valid')

os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)
os.makedirs(valid_folder, exist_ok=True)

# Get the list of file names in the data folder
data_file_names = [file_name for file_name in os.listdir(data_folder) if file_name.endswith('.csv')]

# Shuffle the file names
random.shuffle(data_file_names)

# Calculate the number of files for each set
num_files = len(data_file_names)
train_size = int(0.7 * num_files)
test_size = int(0.15 * num_files)
valid_size = num_files - train_size - test_size

# Split the files into train, test, and validation sets
train_files = data_file_names[:train_size]
test_files = data_file_names[train_size:train_size + test_size]
valid_files = data_file_names[train_size + test_size:]

# Move data files to train folder
for file_name in train_files:
    src_path = os.path.join(data_folder, file_name)
    dest_path = os.path.join(train_folder, file_name)
    shutil.copyfile(src_path, dest_path)

# Move summary files to train folder
for file_name in train_files:
    summary_file_name = os.path.splitext(file_name)[0] + '.txt'
    src_path = os.path.join(summary_folder, summary_file_name)
    dest_path = os.path.join(train_folder, summary_file_name)
    shutil.copyfile(src_path, dest_path)

# Move data files to test folder
for file_name in test_files:
    src_path = os.path.join(data_folder, file_name)
    dest_path = os.path.join(test_folder, file_name)
    shutil.copyfile(src_path, dest_path)

# Move summary files to test folder
for file_name in test_files:
    summary_file_name = os.path.splitext(file_name)[0] + '.txt'
    src_path = os.path.join(summary_folder, summary_file_name)
    dest_path = os.path.join(test_folder, summary_file_name)
    shutil.copyfile(src_path, dest_path)

# Move data files to validation folder
for file_name in valid_files:
    src_path = os.path.join(data_folder, file_name)
    dest_path = os.path.join(valid_folder, file_name)
    shutil.copyfile(src_path, dest_path)

# Move summary files to validation folder
for file_name in valid_files:
    summary_file_name = os.path.splitext(file_name)[0] + '.txt'
    src_path = os.path.join(summary_folder, summary_file_name)
    dest_path = os.path.join(valid_folder, summary_file_name)
    shutil.copyfile(src_path, dest_path)

print("Splitting files into train, test, and valid folders completed!")
```

```python
%%time

import os
import pandas as pd

data_folder = 'data'

# Get the list of CSV files in the data folder
csv_files = [file for file in os.listdir(data_folder) if file.endswith('.csv')]

# Loop through each CSV file
for csv_file in csv_files:
    csv_file_path = os.path.join(data_folder, csv_file)

    # Read the CSV file
    df = pd.read_csv(csv_file_path)

    # Remove the second and fourth columns
    columns_to_remove = [1, 3]  # Index positions of columns to remove
    df.drop(df.columns[columns_to_remove], axis=1, inplace=True)

    # Save the modified dataframe back to CSV
    output_csv_file_path = os.path.join(data_folder, f"modified_{csv_file}")
    df.to_csv(output_csv_file_path, index=False)

print("Columns removed and modified files saved successfully.")
```

```python
import csv
import os

folder_path = 'data'
output_strings = []

# Get the list of CSV files in the folder
file_list = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Iterate over each file in the folder
for file in file_list:
    file_path = os.path.join(folder_path, file)
    file_string = ''

    # Read the CSV file
    with open(file_path, 'r') as file:
        reader = csv.reader(file)

        # Iterate over each row in the CSV file
        for row in reader:
            # Combine the row elements into a single string
            row_string = ': '.join(row)

            # Append the row string to the file string
            file_string += row_string + '\n'

    # Add the file string to the output list
    output_strings.append(file_string)

# Print the output list of strings
#print(output_strings)
```

```python
import os

folder_path = 'summary'
output_summaries = []

# Get the list of .txt files in the folder
file_list = [file for file in os.listdir(folder_path) if file.endswith('.txt')]

# Iterate over each file in the folder
for file in file_list:
    file_path = os.path.join(folder_path, file)

    # Read the contents of the .txt file
    with open(file_path, 'r') as file:
        file_content = file.read()

    # Add the file content to the output list
    output_summaries.append(file_content)

# Print the output list of strings
print(output_summaries)
```
